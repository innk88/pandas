```python
import pandas as pd
from datetime import datetime
df = pd.read_csv ('mission_launches.csv')
df.info()
df.describe()
df.dtypes
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%a %b %d, %Y', errors='coerce')
df['Launch Year'] = df['Date'].dt.year
launch_by_year = df['Launch Year'].value_counts()
max_launch = launch_by_year.max()
year=launch_by_year.idxmax()
print(max_launch, " year:",int(year))
```

```python
success_rate = df['Mission_Status'].value_counts(normalize=True) * 100
print(success_rate)
```

```python
top_countries = df['Location'].value_counts().head(5)
top_organizations = df['Organisation'].value_counts().head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries by Number of Launches')
plt.xlabel('Country')
plt.ylabel('Number of Launches')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
top_organizations.plot(kind='bar', color='lightcoral')
plt.title('Top 5 Organizations by Number of Launches')
plt.xlabel('Organization')
plt.ylabel('Number of Launches')
plt.show()
```
